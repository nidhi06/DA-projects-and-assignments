# Part-1
Find the correlation between: 
Chocolate consumption per capita(kg / head) and Laureates/ 10 million in various countries.
The author had used 23 countries.
He removed the country named Sweden and analysed that the correlation improved.

Here,
The analysis was again done on the same data and the following results were found:

```python
import pandas as pd
df2=pd.read_csv("da_ass5_country23.csv")
print(df2)
```

```python
d=df2['Laureates/ 10 million'].corr(df2['chocolate consumption per capita(kg / head)'])
print("The correlation with all the 23 countries:",d)
```

```python
import statsmodels.api as sm

x=df2['chocolate consumption per capita(kg / head)']
y=df2['Laureates/ 10 million']
x2 = sm.add_constant(x)
est = sm.OLS(y, x2)
est2 = est.fit()
print(est2.summary())
```

p-value = 0.000
This shows high influence of chocolate consumption on the number of noble laureates per country. 

```python
import matplotlib.pyplot as plt
df2.plot(x='chocolate consumption per capita(kg / head)', y='Laureates/ 10 million', style='o')
plt.xlim(0,6)
plt.ylim(0,40)
plt.title('Relationship Between Chocolate Consumption and Laureates ')

#add x and y labels
plt.xlabel('Chocolate Consumption Per Capita( kg/head )')
plt.ylabel('Laureates/ 10 million')
plt.show()
```

```python
import pandas as pd
df3=df2[:22]
d=df3['Laureates/ 10 million'].corr(df3['chocolate consumption per capita(kg / head)'])
print("The correlation when we eliminate Sweden from the list of countries:",d)
```

```python
import matplotlib.pyplot as plt
df3.plot(x='chocolate consumption per capita(kg / head)', y='Laureates/ 10 million', style='o')
plt.xlim(0,6)
plt.ylim(0,40)
plt.title('Relationship Between Chocolate Consumption and Laureates when we eliminate Sweden from the list of countries')

#add x and y labels
plt.xlabel('Chocolate Consumption Per Capita( kg/head )')
plt.ylabel('Laureates/ 10 million')

plt.show()
```
